```python
import pandas as pd
```

Definindo as tabelas e ajustando

```python
produtos_df = pd.read_csv(r"Contoso - Cadastro Produtos.csv", sep=';', encoding='latin1')
vendas_df = pd.read_csv(r'Contoso - Vendas - 2017.csv', sep=';')
clientes_df = pd.read_csv(r'Contoso - Clientes.csv', sep=';', encoding='latin1')
lojas_df = pd.read_csv(r'Contoso - Lojas.csv', sep=';', encoding='latin1')
```

```python
display(vendas_df)
display(produtos_df)
```

Tirando colunas vazias

```python
clientes_df = clientes_df.drop(['Unnamed: 7',	'Unnamed: 8',	'Unnamed: 9',	'Unnamed: 10'], axis=1)
display(clientes_df)
```

Mostrando apenas algumas colunas para dar merge posteriormente 

```python
vendas_df = vendas_df[['ID Produto', 'Data da Venda', 'Quantidade Vendida']]
display(vendas_df)
```

Mostrando apenas algumas colunas para dar merge posteriormente 

```python
produtos_df = produtos_df[['ID Produto', 'ÿNome do Produto']]
display(produtos_df)
```

*Merge* -> novo_dataframe = dataframe1.merge(dataframe2, on='coluna')

```python
vendas_df = vendas_df.merge(produtos_df, on='ID Produto' )
display(vendas_df)
```

Renomear nome de colunas


```python
vendas_df = vendas_df.rename(columns={'ÿNome do Produto':'Nome do Produto'})
```
